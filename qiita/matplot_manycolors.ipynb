```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
import random
print('matplotlib: ', matplotlib.__version__)
print('numpy: ', np.__version__)
print('seaborn: ', sns.__version__)
```

```python
# 各クラスタの点列をランダムに生成する
def generate_2d_points(num_clusters, num_points=100, abs_mu=50, max_std=10):
  # 各クラスタの点列を格納しておく辞書
  d = {}
  for i in range(num_clusters):
    # クラスタごとに，正規分布を用いて点列をランダム生成する
    # 生成に使う正規分布の平均と標準偏差を設定
    mux, muy = np.random.uniform(abs_mu, -abs_mu), np.random.uniform(abs_mu, -abs_mu)
    stdx, stdy = np.random.uniform(max_std, 0), np.random.uniform(max_std, 0)
    # 2次元の正規分布に従って乱数(num_points個の(x,y)の点列)を生成
    d[i] = np.random.normal(loc=[mux, muy], scale=[stdx, stdy], size=(num_points, 2))
  return d

def plot_points(d, label2color):
  fig, ax = plt.subplots(figsize=(8, 8))
  for i in range(len(d)):
    ax.scatter(d[i][:, 0], d[i][:, 1], s=6,
              c=label2color[i], label=i)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(fontsize=12, prop={'size':24}, 
            loc='center left', bbox_to_anchor=(1,0.5), 
            markerscale=10)
  plt.show()

def plot_points_annotated(d, label2color):
  fig, ax = plt.subplots(figsize=(8, 8))
  for i in range(len(d)):
    ax.scatter(d[i][:, 0], d[i][:, 1], s=6,
              c=label2color[i], label=i)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.annotate(i,
              xy=tuple(np.mean(d[i], axis=0)), fontsize=10, color='k',
              bbox={'facecolor':label2color[i], 'edgecolor':'k', 'alpha':0.8})
  plt.show()

def plot_points_tab(d):
  fig, ax = plt.subplots(figsize=(8, 8))
  for i in range(len(d)):
    ax.scatter(d[i][:, 0], d[i][:, 1], s=6,
              c=[matplotlib.cm.tab20(i)]*100, label=i)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend(fontsize=12, prop={'size':24}, 
            loc='center left', bbox_to_anchor=(1,0.5), 
            markerscale=10)
  plt.show()
```

```python
# クラスタ数
num_clusters = 3
# 点列を生成
d = generate_2d_points(num_clusters)
# ラベルと色の対応
label2color = ['r', 'g', 'b']
# プロットを実行
plot_points(d, label2color)
```

```python
def choose_colors(num_colors):
  tmp = list(matplotlib.colors.CSS4_COLORS.values())
  random.shuffle(tmp)
  label2color = tmp[:num_colors]
  return label2color
```

```python
# クラスタ数
num_clusters = 30
# 点列を生成
d = generate_2d_points(num_clusters, abs_mu=600, max_std=40)
# ラベルと色の対応
label2color = choose_colors(num_clusters)
# プロットを実行
plot_points(d, label2color)
plot_points_annotated(d, label2color)
```

```python
# クラスタ数
num_clusters = 20
# 点列を生成
d = generate_2d_points(num_clusters, abs_mu=600, max_std=20)
# ラベルと色の対応
label2color = choose_colors(num_clusters)
# プロットを実行
plot_points_tab(d)
```
